<h1>About AP CSA Exam</h1>

<b>Section I: Multiple Choice </b>

40 Questions | 1 Hour 30 Minutes | 50% of Exam Score

The multiple-choice section includes mostly individual questions, occasionally with 1–2 sets of questions (2 questions per set).
Computational Thinking Practices 1, 2, 4, and 5 are all assessed in the multiple-choice section.

<b>Section II: Free Response</b>

4 Questions | 1 Hour 30 Minutes | 50% of Exam Score

All free-response questions assess Computational Thinking Practice 3: Code Implementation, with the following focus:

Q1: Methods and Control Structures—Students will be asked to write program code to create objects of a class and call methods, and satisfy method specifications using expressions, conditional statements, and iterative statements.

<mark>Q2: Classes </mark>—Students will be asked to write program code to define a new type by creating a class and satisfy method specifications using expressions, conditional statements, and iterative statements.

Q3: Array/ArrayList—Students will be asked to write program code to satisfy method specifications using expressions, conditional statements, and iterative statements and create, traverse, and manipulate elements in 1D array or ArrayList objects.

Q4: 2D Array—Students will be asked to write program code to satisfy method specifications using expressions, conditional statements, and iterative statements and create, traverse, and manipulate elements in 2D array objects.

<b>Free Response notes to keep in mind</b>

1 point penalty 
- Array/collection access confusion ([] get): issue with how you are trying to access certain elements in an array/collection
- Extraneous code that causes SIDE EFFECT (printing to output, incorrect precondition check)
- Local variables used but none declared
- Destruction of persistent data (e.g. changing value referenced by parameter) 
- Void method or constructor that returns a value 

No penalty but common error 
- Mispelling w/no ambiguity 
- Missing ; when structure clearly conveys intent 
- Missing { } where identation clearly conveys intent

<b>STATIC METHODS!</b>

- Cannot be instantiated, associated w/class itself rather than instances of the class. Cannot create objects of a static class
- Use cases: Need to encapsulate utilimity methods, constants, or shared resources that are not tied to any particular instance. Ex. a "Math" class may have static methods for mathematical operations. 
- Use class.method() convention

```java
public class Book 
{
    private String title; 
    private double price; 
    public Book (String bookTitle, double bookPrice)
    {
        // implementation 
    }
    public String getTitle() 
    {
        return title; 
    }

    public String getBookInfo()
    {
        return title + "-" + price; 
    }
}

// Solution: 9 TOTAL POINTS

// 1. Declares class header (must not be private)
public class Textbook extends Book 
{
    private int edition; 

    // 2. Declares constructor header - taking 3 parameters, define type of variable 
    public Textbook(String tbTitle, double tbPrice, int tbEdition) {
        // 3. Call super instructor from book class as the first line w/appropiate parameters 
        super(tbTitle, tbPrice);
        // 4. Declares appropiate private instance variable & uses to appropiate parameter to intialize it 
        edition = tbEdition;
    }

    // 5. Declares AT LEAST 1 required method & all declared headers are correct 
    // 6. getEdition returns value of instance variable 
    public int getEdition() {
        return edition; 
    }

    // 5. 
    // 7. determines whether true or false should be returned based on comparison of book titles and editions (algoritgms)
    public boolean canSubstituteFor(Textbook other) {
        return other.getTitle().equals(getTitle()) && edition >= other.edition; 
    }
    // 5. 
    // 8. getBookInfo calls super.getBookInfo 
    // 9. Constructs info string (return is not assessed for this method) 
    public String getBookInfo() {
        return super.getBookInfo() + "-" + edition; 
    
    }
} 
```
